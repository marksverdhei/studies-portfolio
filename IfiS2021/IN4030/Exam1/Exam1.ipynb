````python
import pandas as pd
blosum62 = pd.read_csv("BLOSUM62.txt", delim_whitespace=True, comment="#")
blosum62
````

````python
def argmax(matrix: List[List[float]]) -> Tuple[int, int]:
    "Exhaustively searches for the highest number in a matrix and outputs it's indices"
    m = len(matrix)
    n = len(matrix[0])
    return max(((i, j) for i in range(m) for j in range (n)), key=lambda t: matrix[t[0]][t[1]])

argmax([[1,2,3]])
````

````python
# Look at https://en.wikipedia.org/wiki/Smith%E2%80%93Waterman_algorithm
def optimal_alignment(seq1: str, seq2: str, 
                      substitution_matrix: pd.DataFrame = blosum62, 
                      extension_penalty: int = 1,
                      opening_penalty: int = 11) -> float:
    """
    Computes the optimal alignment between two nucleotide/amino acid sequences
    using the BLOSUM62 scoring matrix
    
    It uses an affine gap penalty function with gap opening penalty of 11 and gap
    extension penalty of 1.
    """
    
    m = len(seq1)
    n = len(seq2)
    matrix = [[0] * n for i in range(m)]
    
    
    def affine_gap_penalty(n): 
        return opening_penalty + n * extension_penalty

    # TODO: REPLACE WITH ENUMERATE

    
    for i, a1 in enumerate(seq1[1:]):
        for j, a2 in enumerate(seq2[1:]):
            #amino acid 1 and 2
            match_score = substitution_matrix[a1][a2]
            
            penalty = extension_penalty
            
            diag = matrix[i][j] + match_score
            top = max(range(j+1), key=lambda k: matrix[i][k] - affine_gap_penalty(j-k))
            left = max(range(i+1), key=lambda k: matrix[k][j] - affine_gap_penalty(i-k))
            matrix[i+1][j+1] = max(0, diag, top, left)
    
    i, j = argmax(matrix)
    max_score = matrix[i][j]
    
    # Traceback
    align1, align2 = [], []
    
    current_score = max_score

    while current_score > 0:
        directions = {
            "UP": matrix[i-1][j],
            "LEFT": matrix[i][j-1],
            "DIAG": matrix[i-1][j-1]
        }

        direction, current_score = max(directions.items(), key=lambda x: x[1])

        if direction == "UP":
            i -= 1
            align1.append(seq1[i])
            align2.append("-")
        elif direction == "LEFT":
            j -= 1
            align1.append("-")
            align2.append(seq2[j])            
        elif direction == "DIAG":
            i -= 1
            j -= 1
            align1.append(seq1[i])
            align2.append(seq2[j])
        
    align1 = "".join(reversed(align1))
    align2 = "".join(reversed(align2))
    
    return align1, align2, max_score
````
